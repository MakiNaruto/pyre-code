# 🟡 Solution: SwiGLU Activation

Reference solution for the SwiGLU activation function.

$$\text{SwiGLU}(x) = (xW_1) \odot \text{swish}(xW_{\text{gate}}), \quad \text{where} \quad \text{swish}(z) = z \cdot \sigma(z)$$

```python
import torch
```

```python
# ✅ SOLUTION

def swiglu(x, W1, W2, Wgate):
    h = x @ W1
    gate = x @ Wgate
    swish_gate = gate * torch.sigmoid(gate)
    return (h * swish_gate) @ W2
```

```python
# Verify
torch.manual_seed(0)
B, D, H = 2, 8, 16
x = torch.randn(B, D)
W1 = torch.randn(D, H)
W2 = torch.randn(H, D)
Wgate = torch.randn(D, H)

out = swiglu(x, W1, W2, Wgate)
print("Output shape:", out.shape)

gate = x @ Wgate
swish_gate = gate * torch.sigmoid(gate)
print("Gate (swish) sample values:", swish_gate[0, :4])
```
